```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

file_path = r"C:\Users\kimge\OneDrive\문서\Desktop\김 건\가천대학교\2025년 4학년 1학기_시간표\인공지능개론\4_week\wine\wine.csv"
df = pd.read_csv(file_path)
print(df.columns)
print(df)
```

```python
df = df.dropna() # 결측지 제거
df.isnull().sum() 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_corr=df.corr()
plt.figure(figsize=(10,10))
sns.set(font_scale=0.8)
sns.heatmap(df_corr, annot=True, cbar=False) # 히트맵으로 상관 관계 확인
plt.show() # Acl, Nonflavanoid.phenols, Malic.acid 등으로 관련
```

```python
X = df.drop(columns=['Wine'])
y = df['Wine']
print(X.head())
y.value_counts()
print()
Y = pd.get_dummies(y).values # 원-핫 코딩
X = X.values
print(X)
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=0) 
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
#Defining the model 
from keras.models import Sequential
from keras.layers import Dense, Input
from keras.optimizers import SGD,Adam

model = Sequential()
model.add(Input(shape=(13,)))  # Input 객체로 첫 번째 레이어를 수정, X의 개수만큼 입력해야 함(shape를 통해서 구하기)
model.add(Dense(12, activation='relu')) # 12개의 노드를 가진 dense를 추가 
model.add(Dense(10, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(3, activation='softmax')) # categorical_crossentropy사용 -> wine의 분류가 3종류기 때문 
model.compile(optimizer=Adam(learning_rate=0.003), loss='categorical_crossentropy', metrics=['accuracy']) # Adam optimizer에서 lr -> learning_rate로 수정
model.summary()
```

```python
model_history=model.fit(x=X_train, y=Y_train, epochs=30, batch_size=32,validation_data= (X_test,Y_test))
Y_pred = model.predict(X_test)
Y_test_class = np.argmax(Y_test,axis=1)
Y_pred_class = np.argmax(Y_pred,axis=1)
```

```python
accuracy = accuracy_score(Y_test_class, Y_pred_class)
print(f"Accuracy : {accuracy:.4f}")
```

```python
# 훈련 손실과 검증 손실을 추출
loss = model_history.history['loss']
val_loss = model_history.history['val_loss']
epochs = range(1, len(loss) + 1) # 에포크 수를 설정
plt.plot(epochs, loss, 'y', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and Validation Loss for Wine Classification')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
acc =model_history.history['accuracy']
val_acc =model_history.history['val_accuracy']
plt.plot(epochs, acc, 'y', label='Training acc')
plt.plot(epochs, val_acc, 'r', label='Validation acc')
plt.title('Training and Validation Loss for Wine Classification')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(Y_test_class, Y_pred_class, zero_division=1)) # classification_report에서 zero_division 파라미터 사용
print(confusion_matrix(Y_test_class, Y_pred_class)) # confusion_matrix 출력
```
