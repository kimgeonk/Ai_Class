```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras

file_path = r"C:\Users\kimge\OneDrive\문서\Desktop\김 건\가천대학교\2025년 4학년 1학기_시간표\인공지능개론\4_week\heart\heart.csv"
df = pd.read_csv(file_path)
print(df)
```

```python
print(df.columns) # 컬럼명 확인
```

```python
print(df['target'].value_counts()) # 레이블(target) 개수 확인
```

```python
print(df.isnull().sum()) # 결측지 확인 
```

```python
df = df.dropna() # 결측지 제거
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

df_corr=df.corr() # 각 컬럼 별 상관계수를 나타냄 (-1 ~ 1사이로)
plt.figure(figsize=(10,10))
sns.heatmap(df_corr, annot=True, cbar = False) # annot : 상관관계를 나타냄, cbar : 왼쪽에 바를 표시
plt.show()
```

```python
fig = plt.figure(figsize = (15,20)) 
ax = fig.gca() # 현재 활성화되어 있는 축을 가져옴
df.hist(ax = ax) # 모든 컬럼을 대상으로 히스토그램을 생성 
plt.show()
```

```python
X = df.drop(columns = ['target'])
y = df['target']
print(X.head())
print(y.value_counts())
print(y.head())
print(X.shape)
print(y.shape)
```

```python
Y = pd.get_dummies(y).values # 원-핫 엔코딩, numpy로 변환
X = X.values # numpy로 변환
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD,Adam

model = Sequential() # 모델 생성
model.add(Dense(10,input_shape=(13,),activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(6,activation='relu'))
model.add(Dense(2,activation='sigmoid')) # target이 0 or 1로 이루어져 있기 때문에 binary사용
model.compile(Adam(learning_rate=0.003), 'binary_crossentropy', metrics=['accuracy']) # Keras에서는 lr 파라미터가 더 이상 사용되지 않으며, 대신 learning_rate를 사용
model.summary()
```

```python
model_dr = model.fit(X_train, Y_train, epochs=50, batch_size=32,validation_data= (X_test,Y_test)) # 모델 훈련 상태를 추적하는 용도
Y_pred = model.predict(X_test) # 왜 model_dr로 학습하는게 아니지? 
```

```python
from sklearn.metrics import accuracy_score

Y_test_class = np.argmax(Y_test,axis=1) # 원-핫 엔코딩 값중에 가장 큰 값을 반환 
Y_pred_class = np.argmax(Y_pred,axis=1)
accuracy = accuracy_score(Y_test_class, Y_pred_class) # 정확도 test
print(f"Accuracy : {accuracy:.4f}")
```

```python
loss = model_dr.history['loss'] # train 손실 
val_loss = model_dr.history['val_loss'] # test 손실 
epochs = range(1, len(loss) + 1) # 에포크 수를 설정
plt.plot(epochs, loss, 'y', label='Training loss') # plt.plot(x축, y축, '색상', label='범례')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and Validation Loss for target regression')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
acc =model_dr.history['accuracy'] # train 정확도 
val_acc =model_dr.history['val_accuracy'] # test 정확도 
plt.plot(epochs, acc, 'y', label='Training acc')
plt.plot(epochs, val_acc, 'r', label='Validation acc')
plt.title('Training and Validation Loss for target regression')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(Y_test_class, Y_pred_class, zero_division=1)) # percision(정확도), recall(재현율), f1-score(젛확도와 재현율의 평균), support(샘플의 수)
print(confusion_matrix(Y_test_class, Y_pred_class)) # confusion_matrix 출력
```
